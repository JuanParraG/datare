<h3> Predicion valores faltantes con ML

<ul>
<li>Entrenamos modelos para ajsutar los valores de N_clicks
<li> Tres modelos probados: regresion lineal, KNN y Random Forest (mirando un modelo mas robusto) el que mejor metrica tiene es el KNN

```python
#Datos de entrenamiento
train_data_clean_X = pysqldf("""SELECT  julianday(Periodo),
                                        N_Registros_IN,
                                        N_Primer_Deposito_IN,	
                                        Total_Depositado_IN,
                                        N_Personas_Que_Apostaron_IN,	
                                        Total_Apostado,	    
                                        Net_Revenue_IN
                                FROM data_norm 
                                WHERE N_Clicks IS NOT NULL""")

train_data_clean_y = pysqldf("""SELECT N_Clicks	
                                FROM data_norm
                                WHERE N_Clicks IS NOT NULL""")

X_train, X_test, y_train, y_test = train_test_split(train_data_clean_X, train_data_clean_y, test_size=0.2, random_state=42)
```

```python
#Modelo regresion lineal
param_grid = {
    'fit_intercept': [True, False],
}

imputer =LinearRegression()

grid_search = GridSearchCV(imputer, param_grid, scoring='neg_mean_squared_error', cv=5)

grid_search.fit(X_train, y_train)  
predictions = grid_search.predict(X_test)
rf_r2 = r2_score(y_test, predictions)
print(f"linear regression R¬≤ Score: {rf_r2}")
```

```python
#Modelo regresion lineal KNN
param_grid = {
    'n_neighbors': range(1, 20), 
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'] 
}

knn = KNeighborsRegressor()

grid_search_KNN = GridSearchCV(estimator=knn, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5, n_jobs=-1)

grid_search_KNN.fit(X_train, y_train)
predictions = grid_search_KNN.predict(X_test)

knn_r2 = r2_score(y_test, predictions)

print(f"KNN R¬≤ Score: {knn_r2}")
```

```python
# Modelo de Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['auto', 'sqrt'],
    'max_depth': [10, 20, 30, 60],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

imputer = RandomForestRegressor()

rf_grid_search = GridSearchCV(imputer, param_grid=param_grid, cv=5, verbose=0, n_jobs=-1)

rf_grid_search.fit(X_train, y_train.values.ravel())
predictions = rf_grid_search.predict(X_test)

rf_r2 = r2_score(y_test, predictions)

print(f"Random Forest R¬≤ Score: {rf_r2}")
```

<h3>Prescriptiva

Se probara modelos para predecir Cluster de acuerdo a lo analisado con respecto a las variables Total_Depositado_IN Total_Apostado
<ul>
<li>El modelo que se usara es el de KMeans y se eligira la cantidad de cluster de acuerdo a las medidas en inertia silhouette_scores

```python
#Grafico
features = data_norm[['Total_Depositado_IN', 'Total_Apostado']]
inertia = []
silhouette_scores = []
K = range(2, 11) 
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(features, kmeans.labels_))

fig, ax1 = plt.subplots(figsize=(12, 6))

ax2 = ax1.twinx()
ax1.plot(K, inertia, 'bo-', label='Inercia')
ax2.plot(K, silhouette_scores, 'rs-', label='Coeficiente de Silueta')

ax1.set_ylabel('Inercia')
ax2.set_ylabel('Coeficiente de Silueta')

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title('M√©todo del Codo y Coeficiente de Silueta para Determinar el N√∫mero √ìptimo de Clusters')
```

<ul>
<li>La cantidad de clouster optima es 3

```python
#Modelo LSTM
#Datos de entrenamiento RNN LSTM
days_memory = 7  
def create_dataset(X, y, days_memory  = 7  ):
    Xs, ys = [], []
    for i in range(len(X) - days_memory ):
        Xs.append(X[i:(i + days_memory ), :])
        ys.append(y[i + days_memory ])
    return np.array(Xs), np.array(ys)

X_train_LSTM, y_train_LSTM = create_dataset(X_train.values, y_train.values, days_memory )
X_test_LSTM, y_test_LSTM = create_dataset(X_test.values, y_test.values, days_memory )


model = Sequential()
model.add(LSTM(50, return_sequences=True))
model.add(LSTM(25, return_sequences=True))
model.add(LSTM(5))
model.add(Dense(1))
model.compile(loss='mean_squared_error')


model.fit(X_train_LSTM, y_train_LSTM, epochs=500, verbose=0)
train_predict = model.predict(X_train_LSTM)
test_predict = model.predict(X_test_LSTM)

r2_LSTM = r2_score(y_test_LSTM, test_predict)
r2_LSTM_train = r2_score(y_train_LSTM, train_predict)
print(f"Test R¬≤: {r2_LSTM}, Train R¬≤: {r2_LSTM_train}")

# Graficos
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(y_train_LSTM, label='Datos originales (train)')
plt.plot(train_predict, label='Pron√≥stico (train)', color='red')
plt.legend()
plt.title('Predicci√≥n en el conjunto de entrenamiento')

plt.subplot(1, 2, 2)
plt.plot(y_test_LSTM, label='Datos originales (test)')
plt.plot(test_predict, label='Pron√≥stico (test)', color='red')
plt.legend()
plt.title('Predicci√≥n en el conjunto de prueba')

plt.show()
```

```python
# Modelo RandomForestRegressor
rf = RandomForestRegressor(random_state=42)
param_grid = {
    'max_depth': [10, 20],
}

grid_search_rf = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=0, scoring='r2')
grid_search_rf.fit(X_train, y_train.values.ravel())
best_rf = grid_search_rf.best_estimator_

test_predict_rf = best_rf.predict(X_test)
train_predict_rf = best_rf.predict(X_train)

r2_rf = r2_score(y_test, test_predict_rf)
r2_rf = r2_score(y_train, train_predict_rf)
print(f"Test R¬≤: {r2_rf }, Train R¬≤: {r2_rf}")

# Graficos
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(y_train.values, label='Datos originales (train)')
plt.plot(train_predict_rf, label='Pron√≥stico (train)', color='red')
plt.legend()
plt.title('Predicci√≥n en el conjunto de entrenamiento')

plt.subplot(1, 2, 2)
plt.plot(y_test.values, label='Datos originales (test)')
plt.plot(test_predict_rf, label='Pron√≥stico (test)', color='red')
plt.legend()
plt.title('Predicci√≥n en el conjunto de prueba')

plt.show()
```

```python
#Modelo XGBRegressor
xgbr = XGBRegressor(objective='reg:squarederror', random_state=42)
param_grid = {
    'max_depth': [1, 2, 3, 4],
}
grid_search_xgbr = GridSearchCV(estimator=xgbr, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='r2')

grid_search_xgbr.fit(X_train, y_train.values.ravel())
best_xgbr = grid_search_xgbr.best_estimator_
test_predict_xgbr = best_xgbr.predict(X_test)
train_predict_xgbr = best_xgbr.predict(X_train)

r2_xgbr = r2_score(y_test, test_predict_xgbr)
r2_xgbr = r2_score(y_train, train_predict_xgbr)
print(f"Test R¬≤: {r2_xgbr }, Train R¬≤: {r2_xgbr}")

# Graficos
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(y_train.values, label='Datos originales (train)')
plt.plot(train_predict_xgbr, label='Pron√≥stico (train)', color='red')
plt.legend()
plt.title('Predicci√≥n en el conjunto de entrenamiento')

plt.subplot(1, 2, 2)
plt.plot(y_test.values, label='Datos originales (test)')
plt.plot(test_predict_xgbr, label='Pron√≥stico (test)', color='red')
plt.legend()
plt.title('Predicci√≥n en el conjunto de prueba')
```
